## Constellation codes (suffix)

B maps to BeiDou

E maps to Galileo

G maps to GPS

Q maps to QZSS

R maps to GLONASS

## Satellite visibility / status

VSB, VSE, VSG, VSQ, VSR
These are Visibility / satellite status matrices per constellation

Each inner list ≈ satellites tracked at that epoch

Values usually indicate availability, tracking state, or flags

Shape is typically (time × satellites)

## Signal quality

cn0_B1, cn0_B2, cn0_E1, cn0_E2, cn0_G1, …
These are C/N₀ (carrier-to-noise density) in dB-Hz

Higher = cleaner signal

Zeros = satellite not tracked on that band

### Band examples:

B1/B2 maps to BeiDou

E1/E5 maps to Galileo

G1/G2 maps to GPS

## Measurement uncertainty

prStd_B2, prStd_E1, prStd_G1, …
These are Pseudorange standard deviation (meters)

Lower = more reliable range measurement

Used in positioning filters (KF / WLS / PPP)

```python
from pyspark.sql.functions import explode, col

# Example: flatten cn0_B1 column
df_cn0_B1 = spark_df.select(explode(col("cn0_B1")).alias("cn0_B1_inner"))

# Flatten inner array to single numeric rows
df_cn0_B1_flat = df_cn0_B1.select(explode(col("cn0_B1_inner")).alias("cn0_B1_value"))

df_cn0_B1_flat.show(5)
```
